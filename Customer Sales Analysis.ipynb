```python
import matplotlib.pyplot as plt

df.groupby("Product_Category")["Purchase_Amount"].sum().sort_values(ascending = False).plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("City")["Purchase_Amount"].sum().sort_values(ascending = False).plot(kind="bar")

plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("Payment_Method")["Purchase_Amount"].sum().sort_values(ascending = False).plot(kind="pie", autopct = "%.1f%%")

plt.title("Payment Method Distribution")
plt.ylabel("")

plt.show()
```
